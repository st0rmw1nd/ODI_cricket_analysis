```python
# (7) renaming the columns appropriately

df.rename(columns = {'Player':'Name',
                     'Span':'Career_length',
                     'Mat':'Matches',
                     'Inns':'Innings',
                     'Balls':'Balls_bowled',
                     'Wkts':'Wickets',
                     'Ave':'Average',
                     'Econ':'Economy',
                     'SR':'Strike_rate',
                     4:'4_wickets',
                     5:'5_wickets'}, inplace = True)
```

```python
# (8) Finding out how many players played for ICC?

df[["Player_Name", "Country"]] = df['Name'].str.split("(", expand = True)
df['Country'] = df['Country'].str.replace(")", "")
df
```

```python
new_col_sequence = ['Player_Name', 'Country', 'Career_length', 'Matches', 'Innings', 'Balls_bowled', 'Runs', 'Wickets',
       'Average', 'Economy', 'Strike_rate', '4_wickets', '5_wickets']

df = df[new_col_sequence]

# display(df.head())
# df['Country'].value_counts()
```

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
def asia_check(x):
    if "Asia" in x:
        return "Yes"
    else:
        return "No"
```

```python
def africa_check(x):
    if "Afr" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Country'].apply(icc_check)
df['played_for_Asia'] = df['Country'].apply(asia_check)
df['played_for_Africa'] = df['Country'].apply(africa_check)

# display(df.head(10))

print(df['played_for_ICC'].value_counts())
print(df['played_for_Asia'].value_counts())
print(df['played_for_Africa'].value_counts())

df['Country'] = df['Country'].str.replace("ICC/", "")
df['Country'] = df['Country'].str.replace("/ICC", "")
df['Country'] = df['Country'].str.replace("Asia/", "")
df['Country'] = df['Country'].str.replace("Afr/", "")

display(df.head())
```

So, 13 players played for ICC

```python
# (9) Number of different countries present in the dataset

Country_list = df['Country'].value_counts()
print(Country_list)
```

So, there are 11 different countries present in the dataset

```python
# (10 - 11) Finding out which player(s) had played for the longest and shortest period of time?

df[["CL1", "CL2"]] = df['Career_length'].str.split("-", expand = True)
df["CL1"] = pd.to_numeric(df["CL1"])
df["CL2"] = pd.to_numeric(df["CL2"])
df['Period'] = df['CL2']-df['CL1']
del df['CL1']
del df['CL2']
# df.drop(['CL1', 'CL2'])
max(df['Period'])
```

```python
df.loc[df['Period'] == max(df['Period'])]
```

So, SR Tendulkar played for the longest period.

```python
min(df['Period'])
```

```python
df.loc[df['Period'] == min(df['Period'])]
```

So, BKV Prasad, Saeed Ajmal, BAW Mendis and Rashid Khan played for the shortest period.
